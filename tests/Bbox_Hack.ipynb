```python
import matplotlib.pyplot as plt
import random
import string
import sys

sys.path.insert(0, '/home/joel/Desktop/captcha')
print(sys.path)

from captcha.image import ImageCaptcha2, WheezyCaptcha
```

```python
characters = string.digits + string.ascii_uppercase + string.ascii_lowercase
width, height, n_class = 200, 75, len(characters)
generator = ImageCaptcha2(width=width, height=height)

trials = 10

for i in range(trials):
    n_len = random.randint(4,7)
    random_str = ''.join([random.choice(characters) for j in range(n_len)])
    print(random_str)
    img, boxes = generator.generate_image(random_str, bbox=True)
    display(img)
    for i in range(len(boxes)):
        img.paste(0, boxes[i])
    display(img)
```
